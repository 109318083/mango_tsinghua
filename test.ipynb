# Loading test data

```
import numpy as np
from PIL import Image
from tqdm.notebook import tqdm
import pickle 
import matplotlib.pylab as plt # plot module
import pandas as pd

import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
from torchvision import transforms
```

```
if torch.cuda.is_available():
  device = torch.device('cuda:0')
  print('GPU is avalible.')
  print('Working on:', torch.cuda.get_device_name())
else:
  device = torch.device('cpu')
  print('GPU is not avalible.')
  print('Working on CPU')
```

```
with open('/content/drive/MyDrive/Mango/testpkl/test_featureFin002.pkl', 'rb') as f:
    test_feature = pickle.load(f)
```

## dataset

```
class Data(Dataset):
    def __init__(self, x_list):
        self.x_list = x_list
    
    def __getitem__(self, idx):
        x = self.x_list[idx][0]
        return x

    def __len__(self):
        return len(self.x_list)
```

```
test_set = Data(test_feature)
```

```
len(test_set)
```

```
class Net(nn.Module):   
    # Constructor
    def __init__(self,neuron=100, p=0.1):
        super(Net, self).__init__()
        self.neuron=neuron
        self.drop=p

        self.step = nn.Sequential(
            nn.Linear(2000, self.neuron),
            nn.BatchNorm1d(self.neuron),
            nn.ReLU(),
            nn.Dropout(self.drop),

            nn.Linear(self.neuron, self.neuron),
            nn.BatchNorm1d(self.neuron),
            nn.ReLU(),
            nn.Dropout(self.drop),

            nn.Linear(self.neuron, 2),           
        )
    def forward(self, x):
        x = self.step(x)
        return x
```

```
model = Net()
```

## c0 model

```
test_loader = DataLoader(test_set, batch_size=1)
```

```
# Model class must be defined somewhere
model = torch.load('/content/drive/MyDrive/Mango/nn/model_Tc00.pt')
model.eval()
model.to(device)
```

### predict

```
with torch.no_grad():
    pred_c0 = []
    for x in test_loader:
        x = x.to(device)
        z = model(x)
        _,yhat = torch.max(z.data, 1)
        
        #pred_c0.append(yhat.item())
        pred_c0.append(yhat.item())
```
